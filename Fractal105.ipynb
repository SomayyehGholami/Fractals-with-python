```python
# a refers to the name of one of the well-known fractal image
def Possibility(a): 
    po = random.randrange(10000)
    
    if (po <= (frd.iloc[a, 2]*10000)):
        return 1
    if (po <= ((frd.iloc[a, 2] + frd.iloc[a, 3])*10000)):
        return 2
    if (po <= ((frd.iloc[a, 2] + frd.iloc[a, 3] + frd.iloc[a, 4])*10000)):
        return 3  
    if (po <= ((frd.iloc[a, 2] + frd.iloc[a, 3] + frd.iloc[a, 4] + frd.iloc[a, 5])*10000)):
        return 4  
    if (po <= ((frd.iloc[a, 2] + frd.iloc[a, 3] + frd.iloc[a, 4] + frd.iloc[a, 5]+ frd.iloc[a, 6])*10000)):
        return 5     
    else:
        return 6     
   
```
