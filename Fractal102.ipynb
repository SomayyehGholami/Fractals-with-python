```python
# a refers to the name of one of the well-known fractal image
def Possibility(a): 
    po = random.randrange(1000)
    
    if (po <= (frd.iloc[a, 2]*1000)):
        return 1
    if (po <= ((frd.iloc[a, 2] + frd.iloc[a, 3])*1000)):
        return 2
    if (po <= ((frd.iloc[a, 2] + frd.iloc[a, 3] + frd.iloc[a, 4])*1000)):
        return 3  
    if (po <= ((frd.iloc[a, 2] + frd.iloc[a, 3] + frd.iloc[a, 4] + frd.iloc[a, 5])*1000)):
        return 4     
    else:
        return 5     
   
```

```python
x, y = 0, 0 
sdd.setXscale(0.0, 1.0)
sdd.setYscale(0.0, 1.0)
sdd.setCanvasSize(800, 800) 
colors = [RED, GREEN, BLUE, VIOLET, ORANGE]

for i in range(7):
    sdd.clear(sdd.GRAY)
    sdd.setFontSize(s=48)
    sdd.text(0.2, 0.80, frd.iloc[i, 0])
    sdd.setFontSize(s=28)
    sdd.text(0.2, 0.75,'Points Number:')
    sdd.setFontSize(s=36)
    sdd.text(0.2, 0.70,f'({frd.iloc[i, 1]})')
        
    for j in range(frd.iloc[i, 1]):                
        p = Possibility(i)   
        
        x = (frpx.iloc[i, p][0]*x) + (frpx.iloc[i, p][1]*y) + (frpx.iloc[i, p][2])
        y = (frpy.iloc[i, p][0]*x) + (frpy.iloc[i, p][1]*y) + (frpy.iloc[i, p][2])        

        sdd.setPenColor(colors[p-1])    
        sdd.point(x, y)
        sdd.show(0)
    sdd.show(1)
sdd.clear()       
```
